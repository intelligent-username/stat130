The statement highlights a key idea in statistics: when we conduct tests, were trying to understand a whole population, not just the sample we collect data from. To break it down, we use two main concepts: sample statistics and population parameters.

**Sample statistics** are values we calculate from a smaller group (the sample), like the sample mean (denoted as x). In contrast, a **population parameter** refers to a value that describes the entire population, such as the true mean (). While we gather data from our sample (e.g., individual values like x, x), our goal is to make educated guesses about the population () based on that limited information.

This distinction is important because our samples may not perfectly reflect the entire population. Statistical tests help us assess how likely our sample results would be if the actual population parameter () were a certain value (). By interpreting the results of these tests in terms of the population parameters, we can make more informed decisions and conclusions about the larger group based on the data we have.

By understanding this, we recognize the limitations of our sample and the broader implications of our statistical findings.

##### Introduction

In this experiment, we're testing whether STA130 students can discern if milk or tea was poured first in a cup of tea. Out of 80 students, 49 correctly identified the pouring order. We aim to determine if this result suggests a genuine ability to detect the difference or if it could be due to chance.

The code is structured to perform hypothesis testing with both simulated p-values and exact binomial tests. It is reproducible through the use of np.random.seed(), ensuring consistent results across different runs. 

##### Comparison to original Fisher experiment

While inspired by Fisher's original experiment with Dr. Bristol, our study differs in several ways:

**Population**: The population in question is all STA130 students, generalized from the sample we tested. Fisher's experiment was ran with 8 cups, whereas our experiment is ran on 80 cups. We're also testing a general population ability rather than an individual's claim

**Sample**: The sample consists of 80 randomly selected students who were each given one cup of tea and asked to identify whether the milk or tea was poured first. Out of these 80 students, 49 correctly identified the pouring order.

##### Hypotheses

Null Hypothesis (Ho): p = 0.5

The probability of a student correctly guessing the pouring order is 0.5 (random chance).

Alternative Hypothesis (H1): p > 0.5

The probability of a student correctly guessing the pouring order is greater than 0.5.

(Note: Since we're testing whether or not students can intuit if the milk was poured first or not, we're only testing p > 0.5 to see if they're better than random chance)

In simple terms, Ho assumes that the students have no special ability to tell which was poured first, and their answers are just as good as flipping a coin. The test is designed to check whether the observed result (49 correct answers) is something that could easily happen by chance if the students are guessing, or if this result provides evidence that the students may actually have some ability to distinguish the pouring order.



##### Quantitative Analysis

We'll use a simulation approach to test our hypothesis. To do this, we will:

- Generate 10,000 samples of 80 "guesses" (with a 0.5 probability of success).

- Count how often we get 49 or more successes in these simulations.

- Calculate the p-value as the proportion of simulated results that are at least as extreme as our observed result (49 out of 80).The code below will try to find how often 49 students out of 80 guess correctly.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binomtest

np.random.seed(130)

def simulate_experiment(n_simulations=10000, n_students=80, p_null=0.5):
    return np.random.binomial(n=n_students, p=p_null, size=n_simulations)

def plot_distribution(simulated_results, observed_value):
    plt.hist(simulated_results, bins=30, edgecolor='black', alpha=0.7)
    plt.axvline(observed_value, color='red', linestyle='dashed', linewidth=2, label=f'Observed: {observed_value}')
    plt.axvline(np.mean(simulated_results), color='blue', linestyle='solid', linewidth=2, label=f'Mean: {np.mean(simulated_results):.2f}')
    plt.title('Null Distribution with Observed Result')
    plt.xlabel('Number of Correct Guesses')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()

def calculate_p_value(simulated_results, observed_value, tail='greater'):
    if tail == 'greater':
        return np.mean(simulated_results >= observed_value)
    elif tail == 'less':
        return np.mean(simulated_results <= observed_value)
    elif tail == 'two-sided':
        return np.mean(np.abs(simulated_results - np.mean(simulated_results)) >= np.abs(observed_value - np.mean(simulated_results)))
    else:
        raise ValueError("tail must be 'greater', 'less', or 'two-sided'")

n_students = 80
observed_value = 49
null_hypothesis = "The probability of a student guessing correctly is 0.5."
alternative_hypothesis = "The probability of a student guessing correctly is greater than 0.5."

simulated_results = simulate_experiment(n_simulations=10000, n_students=n_students)
plot_distribution(simulated_results, observed_value)

p_value_simulation = calculate_p_value(simulated_results, observed_value, tail='greater')
print(f"Estimated p-value (Simulation): {p_value_simulation:.4f}")

p_value_exact = binomtest(observed_value, n=n_students, p=0.5, alternative='greater').pvalue
print(f"Exact binomial test p-value: {p_value_exact:.4f}")

mean_null = np.mean(simulated_results)
std_null = np.std(simulated_results)
print(f"Null Distribution Mean: {mean_null:.2f}, Standard Deviation: {std_null:.2f}")

# Additional features

# 1. Confidence Interval
confidence_interval = (np.mean(simulated_results) - 1.96 * np.std(simulated_results) / np.sqrt(len(simulated_results)),
                       np.mean(simulated_results) + 1.96 * np.std(simulated_results) / np.sqrt(len(simulated_results)))

print(f"95% Confidence Interval for Null Distribution Mean: {confidence_interval}")

# 2. Power Calculation
def calculate_power(effect_size, n_students, alpha=0.05):
    # Calculate the critical value for the test
    critical_value = np.mean(simulated_results) + np.std(simulated_results) * 1.96 / np.sqrt(n_students)
    
    # Calculate the power
    power = np.mean(np.random.binomial(n=n_students, p=0.5 + effect_size, size=10000) > critical_value)
    return power

effect_size = 0.1
power = calculate_power(effect_size, n_students)

print(f"Power of the test for an effect size of {effect_size}: {power:.4f}")
````
