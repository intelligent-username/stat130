### Question 1


1. **Simple Linear Regression vs. Multiple Linear Regression:**
   - Simple Linear Regression models the relationship between a single independent variable (X) and a dependent variable (Y).
   - Multiple Linear Regression models the relationship between multiple independent variables (X<sub>1</sub>, X<sub>2</sub>, ..., X<sub>n</sub>) and a dependent variable (Y).
   - The benefit of Multiple Linear Regression is that it allows you to analyze the effect of each independent variable on the dependent variable, while controlling for the other variables. This provides a more comprehensive understanding of the relationships in the data.

2. **Continuous Variables vs. Indicator Variables in Simple Linear Regression:**
   - Continuous variables are measurable quantities that can take on any value within a range (e.g., height, weight, temperature).
   - Indicator variables (or dummy variables) are binary variables that take on values of 0 or 1 to represent categorical data (e.g., gender, having a credit card).
   - In Simple Linear Regression, continuous variables are used directly as the independent variable (`X`), while indicator variables are used to represent categorical data.

3. **Behavior Change with Indicator Variables in Multiple Linear Regression:**
   - When an indicator variable is introduced alongside a continuous variable in a Multiple Linear Regression model, the model assumes a different linear relationship for each category represented by the indicator variable.
   - The linear form for Simple Linear Regression is <br><br>
   **Y = β<sub>0</sub> + β<sub>1</sub>X + ε**<br><br>
   while for Multiple Linear Regression with an indicator variable it becomes <br><br>
   **Y = β<sub>0</sub> + β<sub>1</sub>X + β<sub>2</sub>I + ε**<br><br>
   where `I` is the indicator variable.

4. **Effect of Interactions in Multiple Linear Regression:**
   - Introducing an interaction between a continuous variable and an indicator variable in a Multiple Linear Regression model allows the effect of the continuous variable to vary depending on the category represented by the indicator variable.
   - The linear form for this case would be <br><br>
   **Y = β<sub>0</sub> + β<sub>1</sub>X + β<sub>2</sub>I + β<sub>3</sub>(X*I) + ε**<br><br>
   Where `X*I` represents the interaction term.

5. **Multiple Linear Regression with Categorical Variables:**
   - When a Multiple Linear Regression model includes only indicator variables derived from a non-binary categorical variable, the model assumes a different linear relationship for each category.
   - The linear form for this case would be <br><br>
   **Y = β<sub>0</sub> + β<sub>1</sub>I<sub>1</sub> + β<sub>2</sub>I<sub>2</sub> + ... + β<sub>n</sub>I<sub>n</sub> + ε**
   where **I<sub>1</sub>, I<sub>2</sub>, ..., I<sub>n</sub>**<br><br> 
   are the binary indicator variables representing the categories.<br><br>
   - The necessary binary variable encodings for the categorical variable are typically done using techniques like one-hot encoding or dummy coding.

```python
import pandas as pd
import statsmodels.formula.api as smf

# Step 1: Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/CSCS_data_anon.csv"
df = pd.read_csv(url)

# Step 2: Preprocessing the data
# Let's assume we are interested in a binary outcome, e.g., "Feel_connected_yn" (yes/no for feeling socially connected)
# We also assume it is coded as 1 (Yes) and 0 (No). If not, we should convert it.

# Ensuring binary outcome (already provided in binary form in many survey datasets):
df['Feel_connected_binary'] = (df['Feel_connected_yn'] == 'Yes').astype(int)

# Create some example binary and continuous variables for the model:
# Let's choose 'Age' as a continuous predictor and create a binary indicator for 'Gender' (e.g., Female = 1, Male = 0)
df['Female'] = (df['Gender'] == 'Female').astype(int)

# We may also include 'Education_level' as a categorical variable:
# 'C(Education_level)' will be treated as categorical in the formula notation.

# Step 3: Define and fit the logistic regression model using smf (following the additive form example)
# Additive model specification
formula_additive = 'Feel_connected_binary ~ Age + Female + C(Education_level)'
log_reg_additive = smf.logit(formula=formula_additive, data=df).fit()

# Step 4: Print summary of the model
print(log_reg_additive.summary())

# Step 5: Define and fit a synergistic model with an interaction term (e.g., Age * Female interaction)
formula_synergistic = 'Feel_connected_binary ~ Age * Female + C(Education_level)'
log_reg_synergistic = smf.logit(formula=formula_synergistic, data=df).fit()

# Print summary of the synergistic model
print(log_reg_synergistic.summary())
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
```

1. **Train-Test Split**: 
  - Pokemon dataset is evenly split into training and test sets
  - Missing values in "Type 2" are replaced with "None"
  - Most Pokemon aren't legendary
  - Some sort of balance/correlation between attack and defense, as well as between the Special Attack and Special Defensee

---

```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

2. **Basic Model Performance**:
  - Simple model using Attack and Defense to predict HP achieves R² = 0.148
  - Both predictors are statistically significant (p < 0.05)
  - Model explains only 14.8% of HP variance

  ---

```python
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

3. **Model Validation**:
  - In-sample R² = 0.148 matches model fit
  - Out-of-sample R² = 0.212 shows slightly better performance on test data
  - Reasonable generalization despite low overall predictive power

  ---

```python
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
```

4. **Complex Interaction Model**:
  - Includes interactions between Attack, Defense, Speed, Legendary status, Sp. Atk, and Sp. Def
  - Higher in-sample R² = 0.467 but much lower adjusted R² = 0.369
  - Many interaction terms are statistically significant (p < 0.05)
  - Warning of strong multicollinearity (condition number = 1.2e+16)
---

```python
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

5. **Overfitting Assessment**:
  - In-sample R² = 0.467 shows decent training performance
  - Out-of-sample R² = 0.002 reveals severe overfitting
  - Complex model fails to generalize despite higher training R²

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
import plotly.express as px
import plotly.graph_objects as go

# Number of repetitions
reps = 100

# Arrays to store R-squared values for each repetition
in_sample_Rsquared = np.zeros(reps)
out_of_sample_Rsquared = np.zeros(reps)

# Define the linear form based on model3 specification (you can change it to other models if needed)
linear_form = 'HP ~ Attack + Defense + Speed + Q("Sp. Def") + Q("Sp. Atk")'

# Loop over the number of repetitions
for i in range(reps):
    # Random 50-50 train-test split
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, test_size=0.5)

    # Fit the model on the training set
    model_spec = smf.ols(formula=linear_form, data=pokeaman_train)
    model_fit = model_spec.fit()

    # Calculate in-sample R-squared
    in_sample_Rsquared[i] = model_fit.rsquared

    # Calculate out-of-sample R-squared
    y_test = pokeaman_test['HP']
    y_pred_test = model_fit.predict(pokeaman_test)
    out_of_sample_Rsquared[i] = np.corrcoef(y_test, y_pred_test)[0, 1] ** 2

# Create a DataFrame with the collected R-squared values
df = pd.DataFrame({
    "In Sample R-squared": in_sample_Rsquared,
    "Out of Sample R-squared": out_of_sample_Rsquared
})

# Plot the results
fig = px.scatter(
    df, 
    x="In Sample R-squared", 
    y="Out of Sample R-squared",
    labels={"In Sample R-squared": "In-Sample R²", "Out of Sample R-squared": "Out-of-Sample R²"},
    title="Comparison of In-Sample vs Out-of-Sample R-squared"
)

# Add a reference line y = x for comparison
fig.add_trace(
    go.Scatter(
        x=[0, 1], 
        y=[0, 1], 
        mode='lines', 
        line=dict(color='red', dash='dash'),
        name='y = x (Reference Line)'
    )
)

# Display the plot
# fig.show()
fig.show(renderer="png")
```
