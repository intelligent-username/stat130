### Research Question 1

Does the frequency of COVID-19 preventive behaviours (specifically mask-wearing) correlate with self-rated physical health?

#### Variables:
- **COVID_prevention_masks**: Indicates whether individuals are engaging in mask-wearing as a preventive measure (categorical: never, sometimes, often, always)
  
- **WELLNESS_self_rated_physical_health**: Individual's perception of their physical health (continuous scale: 1-10)

#### Analysis Plan:
- **Primary Analysis**: 
  - Spearman's rank correlation analysis between mask-wearing frequency and self-rated physical health
  - Ordinal regression to examine relationships while controlling for confounding variables

- **Secondary Analysis**: 
  - Multiple regression with interaction terms to assess how demographic factors might moderate the relationship
  
- **Chosen Methods**: Data Visualization and Hypothesis Testing
   - Box plots to display self-rated physical health distributions across mask-wearing frequencies
   - Hypothesis testing with α = 0.05 significance level
   - Control variables: age, pre-existing conditions, socioeconomic status

- **Data Preparation**:
   - Multiple imputation for missing values to preserve sample size and reduce bias
   - Verification of monotonic relationship assumption
   - Assessment of multicollinearity among predictors

- **Assumptions**: 
   - Independent observations
   - Monotonic relationship between variables
   - Proportional odds for ordinal regression
   - No significant multicollinearity among predictors

#### Hypotheses:
- H₀: There is no monotonic relationship between mask-wearing frequency and self-rated physical health (ρ = 0)
- H₁: There exists a positive monotonic relationship between mask-wearing frequency and self-rated physical health (ρ > 0)

### Research Question 2

Is there an association between the level of social activity and self-rated mental health among individuals during the COVID-19 pandemic?

#### Variables:
- **SOCIAL_activity_level**: Frequency of social interactions (categorical: low, moderate, high)
  
- **WELLNESS_self_rated_mental_health**: Self-assessed mental health score (continuous scale: 1-10)

#### Analysis Plan:
- **Primary Analysis**: 
  - One-way ANOVA to test for differences in mental health scores across social activity levels
  - Post-hoc Tukey's HSD tests if ANOVA shows significance

- **Secondary Analysis**: 
  - Calculation of 95% confidence intervals for group means
  - Effect size estimation using η² (eta-squared)
  
- **Chosen Methods**: Data Summarization and Statistical Testing
   - Descriptive statistics (mean, median, SD) for each social activity level
   - Confidence intervals for group means
   - ANOVA with post-hoc analysis
   - Significance level: α = 0.05

- **Data Preparation**:
   - Multiple imputation for missing data to maintain statistical power
   - Assessment of normality using Q-Q plots
   - Levene's test for homogeneity of variance
   - Power analysis to ensure adequate sample size

- **Assumptions**:
   - Normal distribution within groups
   - Equal variances across groups
   - Independent observations
   - Missing data at random (MAR)

#### Hypotheses:
- H₀: All group means are equal (μ₁ = μ₂ = μ₃)
- H₁: At least one group mean differs significantly (μᵢ ≠ μⱼ for some i,j)

### Research Question 3

Does the percentage of time spent working remotely influence self-rated productivity?

#### Variables:
- **WORK_remote_percentage**: Percentage of work time spent remotely (continuous: 0-100%)
  
- **WELLNESS_self_rated_productivity**: Self-assessed productivity level (continuous scale: 1-10)

#### Analysis Plan:
- **Primary Analysis**: 
  - Bootstrap analysis of productivity means across remote work percentage quartiles
  - 10,000 bootstrap samples with stratified sampling within quartiles

- **Secondary Analysis**: 
  - Quadratic regression to test for inverted U-shape relationship
  - Residual analysis to validate model fit
  
- **Chosen Methods**: Bootstrapping and Non-linear Regression
   - Generate bootstrap confidence intervals (95%) for mean productivity
   - Quadratic regression to model non-linear relationship
   - Visualization of fitted curve with confidence bands
   - Cross-validation to assess model stability

- **Data Preparation**:
   - Winsorization at 5th/95th percentiles to retain data structure while managing outliers
   - Verification of sufficient sample size per quartile (n > 30)
   - Assessment of reporting bias through sensitivity analysis
   - Test for quartile stability using different cutoff points

- **Assumptions**:
   - Independent observations
   - Sufficient sample size for stable bootstrap estimates
   - No systematic bias in self-reporting
   - Continuous underlying relationship between variables
   - Adequate spread of observations across remote work percentages

#### Hypotheses:
- H₀: No relationship exists between remote work percentage and productivity
- H₁: Productivity follows an inverted U-shaped relationship with remote work percentage, peaking at moderate levels (40-60%)
